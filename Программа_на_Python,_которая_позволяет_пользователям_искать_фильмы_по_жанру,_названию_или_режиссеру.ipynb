Чтобы создать портал, позволяющий пользователям искать фильмы по жанру, названию или режиссеру, необходимо иметь доступ к базе данных фильмов и соответствующей информации, такой как название, режиссер, жанр и т. д. В этом примере мы будем использовать простую базу данных, состоящую из списка фильмов и их атрибутов.

Вот пример программы на Python, которая позволяет пользователям искать фильмы по жанру, названию или режиссеру:

```python
movies = [
    {'название': 'Крестный отец', 'режиссер': 'Фрэнсис Форд Коппола', 'жанр': 'криминал, драма'},
    {'название': 'Побег из Шоушенка', 'режиссер': 'Фрэнк Дарабонт', 'жанр': 'драма'},
    {'название': 'Темный рыцарь', 'режиссер': 'Кристофер Нолан', 'жанр': 'боевик, криминал, драма'},
    {'название': 'Форрест Гамп', 'режиссерr': 'Роберт Земекис', 'жанр': 'драма, романтика'},
    {'название': 'Властелин колец: Братство кольца', 'режиссер': 'Питер Джексон', 'жанр': 'боевик, приключения, драма'},
    {'название': 'Матрица', 'режиссер': 'Лана Вачовски, Лилли Вачовски', 'жанр': 'боевик, научная фантастика'}
]

def search_movies():
    print('Как бы вы хотели искать фильмы?')
    print('1. По жанру')
    print('2. По названию')
    print('3. Режиссер')
    choice = input('Введите свой выбор (1-3): ')

    if choice == '1':
        genre = input('Укажите интересующий вас жанр: ')
        results = [movie for movie in movies if genre.lower() in movie['жанр'].lower()]
        print(f'{len(results)} фильмы найдены:')
        for movie in results:
            print(f"{movie['название']} ({movie['режиссер']}) - {movie['жанр']}")

    elif choice == '2':
        title = input('Введите название фильма, который вы ищете: ')
        results = [movie for movie in movies if title.lower() in movie['title'].lower()]
        print(f'{len(results)} фильмы найдены:')
        for movie in results:
            print(f"{movie['название']} ({movie['режиссер']}) - {movie['жанр']}")

    elif choice == '3':
        director = input('Введите имя интересующего вас директора: ')
        results = [movie for movie in movies if director.lower() in movie['режиссер'].lower()]
        print(f'{len(results)} фильмы найдены:')
        for movie in results:
            print(f"{movie['название']} ({movie['режиссер']}) - {movie['жанр']}")

    else:
        print('Неверный выбор.')

search_movies()
```
