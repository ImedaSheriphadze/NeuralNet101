Вот уникальная программа на **Python**, которая строит график функции, связанной с гипотезой **Коллатца**, которая считается одной из самых сложных нерешенных проблем в математике. Мы будем использовать следующую функцию:

```python
import matplotlib.pyplot as plt

def collatz_sequence(n):
    """Генерирует последовательность Коллатца, начиная с положительного целого числа n."""
    sequence = [n]
    while n != 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        sequence.append(n)
    return sequence

# Main program
if __name__ == "__main__":
    num = int(input("Введите положительное целое число: "))

    sequence = collatz_sequence(num)

    x = range(len(sequence))
    y = sequence

    plt.plot(x, y, marker='o')
    plt.xlabel('Итерация')
    plt.ylabel('Значениеь')
    plt.title('Последовательность Коллатца')
    plt.grid(True)
    plt.show()
```

Теперь опишем каждую команду и ее назначение:

1) **import matplotlib.pyplot as plt**: импортирует модуль **pyplot** из библиотеки **matplotlib**, что позволяет нам создавать графики.

2) **def collatz_sequence(n)**: Эта строка определяет функцию **collatz_sequence**, которая генерирует последовательность Коллатца, начиная с положительного целого числа **n**.

3) **sequence = [n]**: инициализирует список с именем **sequence** с начальным значением **n**.

4)  **while n != 1**: Это запускает цикл, который продолжается до тех пор, пока **n** не станет равным **1**.

5) **if n % 2 == 0**: Проверяет, делится ли **n** на **2**, указывая, что это четное число.

6) **n = n // 2**: если **n** четно, оно делится на **2**.

7) **else**: Если **n** нечетно (не делится на **2**), этот блок выполняется.

8)  **n = 3 * n + 1**: если **n** нечетно, оно умножается на **3** и увеличивается на **1**.

9) **sequence.append(n)**: Текущее значение **n** добавляется к списку последовательностей.

10) **return sequence**: после завершения цикла возвращается список последовательностей, содержащий последовательность Коллатца.

11) **if __name__ == "__main__"**: Это условие гарантирует, что код внутри этого блока запускается только в том случае, если программа запускается напрямую, а не импортируется как модуль.

12) **num = int(input("Введите положительное целое число: "))**: пользователю предлагается ввести положительное целое число и сохранить его в переменной **num**.

13) **sequence = collatz_sequence(num)**: Функция **collatz_sequence** вызывается с заданным пользователем числом, а результирующая последовательность сохраняется в переменной **sequence**.

14) **x = range(len(sequence))**: это создает объект диапазона той же длины, что и список последовательностей. Он представляет собой x-координаты точек на графике.

15) **y = sequence**: список последовательности присваивается переменной **y**. Он представляет координаты **y** точек на графике.

16) **plt.plot(x, y, marker='o')**: эта команда строит график с координатами **x**, заданными **x**, и координатами **y**, заданными **y**. Аргумент **marker='o'** указывает, что каждая точка должна быть отмечена маркером в виде круга.

17) **plt.xlabel('Итерация')**: это устанавливает метку для оси X на «Итерация».

18) **plt.ylabel('Значение')**: это устанавливает метку для оси Y на «Значение».

19) **plt.title('Последовательность Коллатца')**: Это устанавливает заголовок графика на «Последовательность Коллатца».

20) **plt.grid(True)**: включает линии сетки на графике.

21) **plt.show()**: отображает график на экране.
