Демонстрация изображения, обрезанного круглым патчем.

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.cbook as cbook


with cbook.get_sample_data('/content/imeda.png') as image_file:
    image = plt.imread(image_file)

fig, ax = plt.subplots()
im = ax.imshow(image)
patch = patches.Circle((210, 40), radius=90, transform=ax.transData)
im.set_clip_path(patch)

ax.axis('off')
plt.show()
```
