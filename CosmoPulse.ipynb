# **Описание Алгоритма**
Программа создает двумерный визуальный узор с использованием комбинации тригонометрических функций.

Алгоритм состоит из нескольких ключевых шагов:
1. **Инициализация Сетки:** Создается двумерная сетка координат (**x**, **y**) с заданным разрешением.

2. **Генерация Узора:** Используя элементы сетки, код применяет комбинацию синуса и косинуса с учетом радиального расстояния от центра сетки для создания узора.

3. **Маскирование для Анимации:** Узор "**раскрывается**" из центра к краям путем постепенного увеличения радиуса маски с каждым кадром анимации.

4. **Отображение и Анимация:** Полученный узор отображается в виде изображения, и процесс повторяется для создания последовательности кадров, формирующих анимацию.
