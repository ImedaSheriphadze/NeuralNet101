Мы импортируем необходимые библиотеки: pandas для обработки данных, yfinance для извлечения исторических данных об акциях и matplotlib.pyplot для визуализации данных.

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

Мы определяем переменную тикеров, которая представляет собой список тикеров акций для компаний в желаемом портфеле. Вы можете изменить этот список в зависимости от вашего инвестиционного выбора.



Мы указываем переменные start_date и end_date, чтобы определить диапазон времени для выборки исторических данных. Измените эти переменные в соответствии с желаемым периодом времени.

```python
# Шаг 1. Получите данные с онлайн-сайта
# Используйте библиотеку yfinance для получения исторических данных об акциях
tickers = ['PLTR', 'CXMSF', 'PYPL']  # Примерный список биржевых тикеров
start_date = '2020-01-01'  # Дата начала для исторических данных
end_date = '2023-05-09'  # Дата окончания для исторических данных
```

Мы используем функцию yf.download() из библиотеки yfinance, чтобы получить исторические данные об акциях для указанных тикеров и временного диапазона. Функция возвращает DataFrame, содержащий скорректированные цены закрытия акций.

```python
# Получить исторические данные, используя метод загрузки yfinance
data = yf.download(tickers, start=start_date, end=end_date)['Adj Close']
```

Формируем и анализируем портфолио на полученных данных. В этом примере мы рассчитываем доходность, взяв процентное изменение данных и отбросив все недостающие значения.

Мы определяем переменную portfolio_weights, которая представляет веса, присвоенные каждой акции в портфеле. Измените эти веса в зависимости от желаемого распределения.

```python
# Шаг 2: Выполните формирование и анализ портфеля
# Рассчитать доходность и другие показатели портфеля
returns = data.pct_change().dropna()
portfolio_weights = [0.4, 0.3, 0.3]  # Пример веса портфеля
```

Мы рассчитываем доходность портфеля, умножая доходность акций на их соответствующие веса и суммируя их.

Вы можете рассчитать дополнительные показатели эффективности портфеля, такие как совокупная доходность, стандартное отклонение, коэффициент Шарпа и т. д., исходя из ваших требований.

```python
# Рассчитать доходность портфеля
portfolio_returns = returns.dot(portfolio_weights)

# Расчет показателей эффективности портфеля, таких как совокупная доходность, стандартное отклонение, коэффициент Шарпа и т. д..
```

Мы визуализируем эффективность портфеля, рисуя доходность портфеля с помощью matplotlib.pyplot. Мы указываем размер фигуры, отмечаем доходность портфеля по оси Y относительно соответствующих дат по оси X и добавляем метки и линии сетки для лучшей визуализации.

```python
# Шаг 3. Визуализируйте эффективность портфеля
# График доходности портфеля
plt.figure(figsize=(10, 6))
plt.plot(portfolio_returns.index, portfolio_returns, color='blue')
plt.title('Портфель возвращается')
plt.xlabel('Дата')
plt.ylabel('Возвращает')
plt.grid(True)
plt.show()
```

Следуя этим командам, программа извлекает исторические данные об акциях из онлайн-источника, выполняет формирование и анализ портфеля и визуализирует доходность портфеля.
