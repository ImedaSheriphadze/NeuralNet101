Чтобы создать уникальную программу, которая строит базовую схему вычислительной системы, мы можем использовать **Python** и библиотеку **networkx** для графического представления. Мы создадим простой пример вычислительной системы с такими компонентами, как **ЦП**, **ОЗУ**, **хранилище** и **устройства ввода-вывода**, а затем отобразим график на экране с помощью **networkx** и **matplotlib**.

Во-первых, убедитесь, что у вас установлены библиотеки networkx и **matplotlib**. Вы можете установить их с помощью **!pip**, если вы еще этого не сделали:

```python
import networkx as nx
import matplotlib.pyplot as plt

def build_computing_system_graph():
    # Создать пустой график
    G = nx.DiGraph()

    # Добавьте узлы для ЦП, ОЗУ, хранилища и устройств ввода-вывода.
    G.add_node("CPU", color="blue")
    G.add_node("RAM", color="green")
    G.add_node("Storage", color="yellow")
    G.add_node("I/O Device", color="red")

    # Добавьте ребра для представления соединений между компонентами
    G.add_edge("CPU", "RAM")
    G.add_edge("CPU", "Storage")
    G.add_edge("CPU", "I/O Device")
    G.add_edge("RAM", "Storage")
    G.add_edge("RAM", "I/O Device")
    G.add_edge("Storage", "I/O Device")

    return G

def display_computing_system_graph(graph):
    # Получить цвета узлов
    node_colors = [graph.nodes[node]['color'] for node in graph.nodes]

    # Нарисуйте график
    pos = nx.spring_layout(graph, seed=42)
    nx.draw(graph, pos, with_labels=True, node_color=node_colors, node_size=1000, font_size=10)

    # Показать сюжет
    plt.title("Базовая вычислительная система")
    plt.show()

if __name__ == "__main__":
    computing_system_graph = build_computing_system_graph()
    display_computing_system_graph(computing_system_graph)
```

В этой программе мы определяем функцию **build_computing_system_graph()** для создания графа вычислительной системы с узлами, представляющими такие компоненты, как **ЦП**, **ОЗУ**, **хранилище** и **устройства ввода-вывода**. Мы также определяем ребра для представления связей между этими компонентами. Цвета узлов задаются для визуального различения различных компонентов.

Функция **display_computing_system_graph(graph)** использует **networkx** и **matplotlib** для отображения графика на экране. Узлы выложены с использованием алгоритма пружинной компоновки для более четкой визуализации.
