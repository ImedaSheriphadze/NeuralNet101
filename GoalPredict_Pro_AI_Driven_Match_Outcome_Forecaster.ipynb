Эта программа на **Python** использует методы машинного обучения и нейронных сетей для прогнозирования исхода футбольного матча между двумя командами – **Арсеналом** и **Кристал Пэласом**, который проводится: **20.01.2024 16:30**. Она включает в себя несколько этапов, включая подготовку и нормализацию данных, создание и обучение модели нейронной сети, оценку точности модели и визуализацию результатов.

Вот подробное объяснение каждой части кода:


```python
# Импорт библиотек:
from keras.models import Sequential
from keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt

# Подготовка и нормализация данных
data = np.array([[10, 7, 2, 1, 22, 10, 23, 4],  # Арсенал
                 [10, 3, 3, 4, 11, 15, 12, 12]]) # Кристал Пэлас
labels = np.array([1, 0])

scaler = StandardScaler()
data_normalized = scaler.fit_transform(data)

# Разделение на обучающую и тестовую выборку
X_train, X_test, y_train, y_test = train_test_split(data_normalized, labels, test_size=0.2, random_state=42)

# Создание модели нейронной сети
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=8))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

# Компиляция модели
model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

# Обучение модели
history = model.fit(X_train, y_train, epochs=100, batch_size=1, verbose=0)

# Оценка точности модели на тех же данных, так как тестовый набор отсутствует
scores = model.evaluate(X_train, y_train, verbose=0)
print("\nТочность: %.2f%%" % (scores[1]*100))

# Визуализация результатов обучения
plt.plot(history.history['accuracy'], label='Тренировочная точность')
plt.title('Точность модели')
plt.ylabel('Точность')
plt.xlabel('Эпоха')
plt.legend()
plt.show()

import matplotlib.pyplot as plt

# Создание визуализации
teams = ['Арсенал', 'Кристал Пэлас']
probabilities = [prob_arsenal, prob_crystal_palace]

plt.figure(figsize=(10, 6))
plt.bar(teams, probabilities, color=['red', 'blue'])
plt.xlabel('Команды')
plt.ylabel('Вероятность победы (%)')
plt.title('Прогноз победы в матче Арсенал против Кристал Пэлас')
plt.ylim(0, 1)
plt.show()

```
